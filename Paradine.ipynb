#Kickoff Check

```python
kickoff = pd.merge(alpha_reviews,beta_fixtures,how='inner',on=['canonical_event_id'])
kickoff = kickoff[['canonical_event_id','kick_off_x','kick_off_y']]

def kickoff_check(time_1, time_2):
  check = 0
  if time_1 != time_2:
    check = 1
  return check

kickoff['check'] = kickoff.apply(lambda x: kickoff_check(x.kick_off_x,x.kick_off_y), axis=1)
kickoff
```

#Check for any games that went drastically against Beta odds expectations

```python
def jackpot_checker(odds_A, odds_B, score_A, score_B):
  jackpot = 0
  goal_diff = abs(score_A - score_B)
  odds_diff = abs(odds_A - odds_B)
  if odds_diff >= 3 and goal_diff >= 1:
    jackpot = 1
  return jackpot

odds_check = pd.merge(beta_market,beta_fixtures,how='inner',on=['canonical_event_id'])
odds_check = odds_check[['canonical_event_id','feature_a','feature_b', 'team1_goals', 'team2_goals']]

odds_check['jackpot'] = odds_check.apply(lambda x: jackpot_checker(x.feature_a, x.feature_b, x.team1_goals, x.team2_goals), axis=1)

odds_check
```

```python
odds_check = odds_check[odds_check.jackpot == 1]
odds_check
```
